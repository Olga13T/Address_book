```python
import numpy as np
import pandas as pd
```

```python
fruits = {
    'apples': 101,
    'oranges': 100,
    'plums': 3000}
print(fruits.values()) 
def get_keys(def_dict:dict,keys: list)->dict:
    res = {}
    for i in keys:
        res[i] = def_dict[i]
    return res
```
